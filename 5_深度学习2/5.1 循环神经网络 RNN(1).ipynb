## RNN模型

<span class="motutor-highlight motutor-id_9eugava-id_1s91v6b"><i></i>RNN</span>如下图所示，其输入可以是任意长度，特别适合对顺序敏感的序列问题（时序数据）进行建模。

在RNN中利用循环（上一时刻）的Hidden States，<span class="motutor-highlight motutor-id_9eugava-id_mozeb5h"><i></i>实现“记忆”</span>

![](https://imgbed.momodel.cn/rnn1.png)




![](https://imgbed.momodel.cn/rnn2.png)



*   对于 $t$ 时刻（位置）的输入词来说，首先把它转换为词向量 $e^{(t)}$ ，作为RNN的输入；
*   然后对于隐藏层，它的输入来自两部分，一部分是时刻的输入 $W_ee^{(t)}$，另一部分是上一刻的隐状态 $W_hh^{(t-1)}$，这两部分组合起来再做一个非线性变换，得到当前层的隐状态 $h^{(t)}$；
*   最后，隐状态再接一个 softmax 层，得到该时刻的输出概率分布 $\hat{\boldsymbol{y}}^{(t)}$。

RNN相比于固定窗口的神经网络，其<span class="motutor-highlight motutor-id_mv1zzec-id_qm662ur"><i></i><b>优势</b></span>是：

*   可以处理 **任意长度** 的输入
*   步骤 t 的计算可以感知很久**以前的状态**
*   模型大小是固定的，**模型大小不会** 随着输入的增加而增加
*   所有参数 $W_e,W_h,U,b_1,b_2$ 都是**共享**的

**不足**：

*   递归计算速度 **慢**，因为后续状态需要用到前面的状态，是串行的，难以并行计算
*   虽然理论上t时刻可以感知很久以前的状态，但实际上很难，因为梯度消失的问题（长期记忆问题）
