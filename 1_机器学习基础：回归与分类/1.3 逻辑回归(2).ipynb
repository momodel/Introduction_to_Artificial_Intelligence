**<span class="motutor-highlight motutor-id_iia2jsq-id_41rnnne"><i></i>损失函数</span>**

$$ L=-\frac{1}{m} \sum_{i=1}^m (y_i \log f(x_i) + (1-y_i) \log (1-f(x_i))$$

if $y=1$, $L = - \log (f(x))$
