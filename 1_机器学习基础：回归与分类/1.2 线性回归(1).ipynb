# 线性回归

<span class="motutor-highlight motutor-id_ucntyht-id_ve6k9t2"><i></i>线性回归：</span>
1. $y=ax+b$
2. $y=ax^2 + bx + c$
3. $y=ax_1 + bx_2 + c$ （多元线性回归）


假设选择$y=ax+b$

给定$m$个训练数据$D=\{(x_i, y_i)| i=1,...,m\}$

选择损失函数$L= \frac{1}{m} \sum_{i=1} ^m (y_i - (ax_i + b))^2$  (*Mean Squared Error - MSE*)

找到一组$a$和$b$，使$L$最小



解析解: $\frac{\partial L}{\partial \theta} = 0$, 最小二乘法等

优化方法：**梯度下降**
