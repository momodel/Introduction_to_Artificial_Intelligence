## <span class="motutor-highlight motutor-id_3zo9hze-id_5kyk53j"><i></i>用 Sklearn 实现逻辑回归</span>

```python
import numpy as np
# 引入逻辑回归模型
from sklearn.linear_model import LogisticRegression
```

```python
X = np.array([[82, 62], [79, 59], [81, 63], [72, 68], [73, 70], [74, 68]])
y = np.array([0, 0, 0, 1, 1, 1])
```

```python
# 模型训练
clf = LogisticRegression(random_state=0).fit(X, y)
```

```python
# 预测
clf.predict(np.array([[76, 72]]))
```

```python
clf.score(X, y)
```
