### **<span class="motutor-highlight motutor-id_33d84hw-id_c1d8sji"><i></i>损失函数</span>**

Cost function **$C$**, Loss Function **$L$**

测量：预测的$y^*$与实际的$y$之间的差别

### **<span class="motutor-highlight motutor-id_6d9t7cm-id_l35n79m"><i></i>模型训练 即 损失函数优化</span>**

损失函数是关于什么的函数？

找到一种$\theta$, 使得$y^* = f_\theta(x)$ 与 $y$ 之间差别最小（即 $L(y^*, y)$最小）
