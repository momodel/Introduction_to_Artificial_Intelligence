## <span class="motutor-highlight motutor-id_1ua4zgt-id_0p6rv6g"><i></i>用 Sklearn 实现线性回归</span>

导入所需的库，选择模型


```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

<span class="motutor-highlight motutor-id_lhkwj0w-id_8oru17l"><i></i>准备数据</span>

```python
#准备数据
X = np.array([[5], [10], [12], [13], [15]])
y = np.array([20, 15, 8, 6, 4])
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y,  color='black')
plt.show()
```

<span class="motutor-highlight motutor-id_nu3hpxu-id_11mjnde"><i></i>模型训练</span>

```python
# 模型训练，自动调用梯度下降
reg = LinearRegression().fit(X, y)
```

```python
# a
reg.coef_
```

```python
# b
reg.intercept_
```

<span class="motutor-highlight motutor-id_umuf899-id_lfcpwqq"><i></i>评价模型</span>

```python
# 评分（1是最高）
reg.score(X, y)
```

模型预测

```python
# 预测
reg.predict(np.array([[3]]))
```

```python
plt.scatter(X, y,  color='black')

x_test = np.linspace(0, 15, 100)
y_test = reg.predict(x_test[:, np.newaxis])

plt.plot(x_test, y_test, color='blue', linewidth=3)


plt.show()
```
