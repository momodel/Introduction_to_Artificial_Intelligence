通过一段简单的代码来理解<span class="motutor-highlight motutor-id_zr78g4l-id_tu36b1i"><i></i>欠拟合与过拟合</span>问题：

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 用身高X 预测寿命y
X = np.array([[1.83], [1.77], [1.67], [1.75], [1.72], [1.71], [1.80]])
y = np.array([74, 71, 78, 74, 69, 80, 77])

reg = LinearRegression().fit(X, y)

reg.score(X, y)
```

<span class="motutor-highlight motutor-id_6qwr0yo-id_9b9cgbf"><i></i>将模型变得更复杂</span>：

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X1 = np.array([1.83, 1.77, 1.67, 1.75, 1.72, 1.71, 1.80])
X2 = [x**2 for x in X1]
X3 = [x**3 for x in X1]
X4 = [x**4 for x in X1]

y = np.array([74, 71, 78, 74, 69, 80, 77])


X = np.stack((X1, X2, X3, X4))
X = np.transpose(X)

reg = LinearRegression().fit(X, y)

reg.score(X, y)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X1 = np.array([1.83, 1.77, 1.67, 1.75, 1.72, 1.71, 1.80])
X2 = [x**2 for x in X1]
X3 = [x**3 for x in X1]
X4 = [x**4 for x in X1]
X5 = [x**5 for x in X1]
X6 = [x**6 for x in X1]

y = np.array([74, 71, 78, 74, 69, 80, 77])


X = np.stack((X1, X2, X3, X4, X5, X6))
X = np.transpose(X)

reg = LinearRegression().fit(X, y)

reg.score(X, y)
```
