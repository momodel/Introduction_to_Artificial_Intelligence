# 岭回归和Lasso回归

## 岭回归 - <span class="motutor-highlight motutor-id_hhjdz6z-id_2nnrgdj"><i></i>L2正则化</span>
$L = \frac{1}{2m}\sum_{i=1}^m(f(x_i)-y_i)^2 + \lambda \sum_{j=1}^n \theta_j^2$


## Lasso回归 - <span class="motutor-highlight motutor-id_hhjdz6z-id_kalwanv"><i></i>L1正则化</span>
$L = \frac{1}{2m}\sum_{i=1}^m(f(x_i)-y_i)^2 + \lambda \sum_{j=1}^n |\theta_j|$

```python
# 岭回归如何减少过拟合

import numpy as np
from sklearn.linear_model import Ridge

X1 = np.array([1.83, 1.77, 1.67, 1.75, 1.72, 1.71, 1.80])
X2 = [x**2 for x in X1]
X3 = [x**3 for x in X1]
X4 = [x**4 for x in X1]
X5 = [x**5 for x in X1]
X6 = [x**6 for x in X1]

y = np.array([74, 71, 78, 74, 69, 80, 77])


X = np.stack((X1, X2, X3, X4, X5, X6))
X = np.transpose(X)

clf = Ridge(alpha=1.0)
clf.fit(X, y)

print(clf.score(X, y))
print(clf.coef_)
```

```python
# Lasso如何减少过拟合

import numpy as np
from sklearn import linear_model

X1 = np.array([1.83, 1.77, 1.67, 1.75, 1.72, 1.71, 1.80])
X2 = [x**2 for x in X1]
X3 = [x**3 for x in X1]
X4 = [x**4 for x in X1]
X5 = [x**5 for x in X1]
X6 = [x**6 for x in X1]

y = np.array([74, 71, 78, 74, 69, 80, 77])


X = np.stack((X1, X2, X3, X4, X5, X6))
X = np.transpose(X)

clf = linear_model.Lasso(alpha=0.1)
#clf = linear_model.Lasso(alpha=0.04)
clf.fit(X, y)

print(clf.score(X, y))
print(clf.coef_)
```
