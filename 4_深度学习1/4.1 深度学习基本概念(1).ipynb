## <span class="motutor-highlight motutor-id_cwa4jbz-id_y2watib"><i></i>为什么我们需要更深的网络</span>



### Universality 

**Universal Approximation Theorem**: 通用近似定理（或称万能近似定理）指出人工神经网路近似任意函数的能力

神经网路可以用来<span class="motutor-highlight motutor-id_m3emi65-id_q85ucap"><i></i>近似任意的复杂函数</span>，并且可以达到任意近似精准度。

*它并没有告诉我们如何选择神经网络参数（权重、神经元数量、神经层层数等等）来达到我们想近似的目标函数。*

乔治·西本科于1989年证明了单一隐藏层、任意宽度、并使用S函数作爲激励函数的前馈神经网路的通用近似定理。科特·霍尼克（英语：Kurt Hornik）在1991年证明 ，激励函数的选择不是关键，前馈神经网路的多层神经层及多神经元架构才是使神经网络有成为通用逼近器的关键。

>Cybenko, G. (1989) "Approximations by superpositions of sigmoidal functions", Mathematics of Control, Signals, and Systems, 2(4), 303–314. doi:10.1007/BF02551274

>Kurt Hornik (1991) "", Neural Networks, 4(2), 251–257. doi:10.1016/0893-6080(91)90009-T
