<h2><span class="motutor-highlight motutor-id_lucbtpw-id_1x9er04"><i></i>训练中卷积层的维度变化</span></h2>
下面我们从一个示例来过一遍整个过程，视频中使用的是 Keras，读者并不需要学习这个框架，此处只是为了演示我们最初的图片矩阵在 CNN 中的维度变化。

假设我们图片是一张黑白的 $28 \times 28$ 大小的图片，如下图所示。

![](https://imgbed.momodel.cn/20201105114859.png)

第一次卷积的输入是原始图片，图片是黑白的，所以第一个维度为 1（如果是 RGB 就是 3），所以 `input_shape=(1, 28, 28)` , 第一层卷积层有 25 个 $3 \times 3$ 大小的 filter，所以卷积操作为 `Convolution2D(25, 3, 3)`。

在进行完第一次卷积之后，我们得到的输出结果是一个 $25 \times 26 \times 26$ 的张量：25 个 filter对应第一个维度，每个 $3 \times 3$ 的 filter 要对 $28 \times 28$ 的图片上内积 $26 \times 26$ 次，所以后两个维度是 $26 \times 26$。

然后我们进行池化操作，我们把每个 $26 \times 26$ 的矩阵分成许多个 $2 \times 2$ 的小矩阵，然后从每个小矩阵中选出一个最大的值，对应的操作是 `MaxPooling2D((2, 2))`，这时候我们就完成了第一卷积+池化的操作。得到的结果为 $25 \times 13 \times 13$ 的张量，可以看成是 25 张 $13 \times 13$ 的小图片，然后对这 25 张图片进行第二次卷积+池化，如下图所示。

![](https://imgbed.momodel.cn/20201105120239.png)

在第二次卷积的时候要注意，我们输入的为 `Convolution2D(50, 3, 3)`，这里并不是说我们每个 filter 也是 $3 \times 3$ 的大小，而是 $25 \times 3 \times 3 = 225$ 的大小的 filter，50 表示有 50 个 filter，因为在上一步的操作中，我们得到的结果是 $25 \times 13 \times 13$ 的，表示有 25 个 channel，框架会自动帮我们处理中间过程，所以我们传入的参数只需 `(50, 3, 3)`。

到这一步，卷积的操作就基本结束了，我们剩下要做的事情就是把 $50 \times 5 \times 5$ 的张量变成 1250 大小的向量，然后输入到一个全连接神经网络，用我们之前学过的方法进行训练就行了。
