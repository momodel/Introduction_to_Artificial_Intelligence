## 神经网络 与 人工神经网络

<span class="motutor-highlight motutor-id_7fxxvpb-id_006webs"><i></i>“飞鸟派”</span>：如果我们想要学飞翔，就得向“飞鸟”来学习。

简单来说，“飞鸟派”就是“仿生派”，即把进化了几百万年的生物，作为“模仿”对象，搞清楚原理后，再复现这些对象的特征。

神经网络模拟人脑神经元的连接来达到学习功能，通过逐层抽象将输入数据逐层映射为概念等高等语义。

其实现在所讲的神经网络包括深度学习，都在某种程度上，属于“飞鸟派”——它们在模拟大脑神经元的工作机理。


一个<span class="motutor-highlight motutor-id_z2up9ve-id_zttrwpx"><i></i>神经元</span>通常具有多个树突，主要用来接受传入信息；而轴突只有一条，轴突尾端有许多轴突末梢可以给其他多个神经元传递信息。  
轴突末梢跟其他神经元的树突产生连接，从而传递信号。这个连接的位置在生物学上叫做“突触”。
