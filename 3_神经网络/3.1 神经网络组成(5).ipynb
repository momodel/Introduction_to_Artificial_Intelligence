## <span class="motutor-highlight motutor-id_m1hicxg-id_gvzgump"><i></i>前序网络及其中的操作</span>



神经网络：相互连接形成一个无环图的神经元集合。

神经网络通常由不同的神经元层组织而成，常见的层类型有：
+ 全连接层
+ 卷积层
+ 池化层
+ Dropout层
+ RNN层
+ ......

<span class="motutor-highlight motutor-id_w865ynp-id_bqvyepi"><i></i>多层感知机（神经网络）</span>比简单的单感知机模型从结构上来看要复杂的多。

<img src="http://imgbed.momodel.cn//20200103111837.png" width=400>

+ 输入层和输出层之间存在若干隐藏层。
+ 每个隐藏层中包含若干神经元。
+ 除了输出层，每一层的神经元都与下一层的所有神经元连接。



+ 前馈神经网络接收一个输入 $x$ ，并产生一个输出 $\hat{y}$ 信息正向流过网络，称为**正向传播(forward propagation)**

+ 训练过程中：正向传播可以一直继续下去直到产生一个标量代价
$J(\theta)$ .
