# 反向传播



<span class="motutor-highlight motutor-id_vvuguax-id_8dx3jwd"><i></i>反向传播</span>的主要任务，就是求解出神经网络模型中权重层对代价函数的**偏导数**，然后我们就可以使用梯度下降法来更新模型权重。

**反向传播** 和梯度下降的结合，就可以训练任何复杂的神经网络。

+ **核心问题**：给定某个函数 ，计算f在点x处的梯度$\epsilon{f(x)}$

+ 在神经网络中， 对应于**损失函数**，而输入x 则是由**训练数据**和神经网络**权重参数**构成

+ 反向传播：通过递归地应用链式法则来计算表达式梯度的一种方法
